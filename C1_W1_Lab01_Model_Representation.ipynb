# Goal
Implement Linear Regression Model $f_{w,b}$ with one feature.


```python
#Library for computing
import numpy as np
#Library for plotting
import matplotlib.pyplot as matplot
```

```python
#x_train is the input variable (size in 1000 square feet)
#y_train is the target variable (price in 1000s of dollars)
x_train = np.array([1.0,2.0,4.0])
y_train = np.array([300.0,600.0,1200.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape = {x_train.shape}")
print(f"y_train.shape = {y_train.shape}")
```

```python
#size of training set can be found using len function
m = print(f"number of taining examples m = {len(x_train)}")
```

```python
#plotting the training set
matplot.scatter(x_train,y_train,marker='*',c='g')
matplot.title("Housing prices")
matplot.ylabel("Price (in 1000s of dollars)")
matplot.xlabel("size (in 1000 sqfeet)")
matplot.show()
```

# Linear regression model formula:
$ f_{w,b}(x^{(i)}) = wx^{(i)} + b $
    

```python
w1 = 100
b1 = 100
print(f"w1 for model = {w1}")
print(f"b1 for model = {b1}")
```

```python
def compute_linear_regression_model_output(x,w,b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b;
    return f_wb
```

```python
predictedf_wb1 = compute_linear_regression_model_output(x_train,w1,b1)

#plot the model prediction
matplot.plot(x_train, predictedf_wb1, c='b', label='Model Prediction')
#plot the actual data points
matplot.scatter(x_train, y_train, c='r', label='Actual values')

matplot.title("Housing Prices")
matplot.ylabel("Price (in 1000s of dollars)")
matplot.xlabel("size (in 1000 sqfeet)")
matplot.legend()
matplot.show()
```

```python
w2 = 200
b2 = 100
print(f"w2 for model = {w2}")
print(f"b2 for model = {b2}")
```

```python
predictedf_wb2 = compute_linear_regression_model_output(x_train,w2,b2)

#plot the model prediction
matplot.plot(x_train, predictedf_wb2, c='b', label='Model Prediction')
#plot the actual data points
matplot.scatter(x_train, y_train, c='r', label='Actual values')

matplot.title("Housing Prices")
matplot.ylabel("Price (in 1000s of dollars)")
matplot.xlabel("size (in 1000 sqfeet)")
matplot.legend()
matplot.show()
```

# Prediction
We can predict the price of a house given the size using the model

```python
w = 200
b = 100
x_i = 3.0
cost_3000sqft = w * x_i + b
print(f"Estimated cost of 3000 sq feet will be = {cost_3000sqft}")
```
